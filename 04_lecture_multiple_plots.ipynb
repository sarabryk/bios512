```r
# alternatively add data and mapping in the geom

p = ggplot() +
    geom_point(data = mpg, mapping = aes(x = displ, y = hwy))

p
```

```r
# How to change the color of all the points (i.e. not map them to a column in the data)?

p = ggplot(mpg, aes(x = displ, y = hwy))

p = p + geom_point(color = 'firebrick', alpha = 0.5)

p
```

```r
# What other chart properties can be "mapped?"
# X, Y, color...
# Hint: # https://r4ds.had.co.nz/data-visualisation.html#fig:shapes

p = ggplot(mpg, aes(x = cty, y = hwy))

p = p + geom_point(shape = 22, fill = 'pink', color = 'black')

p
```

# Facets

## facet_wrap

```r
# Let's use a different data set

iris |> head()
```

```r
# this will make each species a different color! 
p = ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species))

# what does geom_point do? 
p = p + geom_point() 

p = p + scale_color_brewer(type = "qual", palette = "Set2")

p
```

```r
options(repr.plot.width = 8, repr.plot.height = 6)

p = ggplot(iris, aes(x = Sepal.Length, 
                     y = Sepal.Width, 
                     color = Species))

p = p + geom_point() 

p = p + scale_color_brewer(type = "qual", 
                           palette = "Set2")

# FACET_WRAP COMMAND 
# The facet_wrap command uses a variable that you specify (Species) to break out the data, by species, into individual plots.  
# That grid of charts that you create will then have 2 rows. (you can adjust the rows to make the chart space taller/wider.)
p = p + facet_wrap(~Species, nrow = 2)
    # we put the ~Species - the ~ doesn't mean anything it's just part of the command language.  

p
```

```r
options(repr.plot.width=4, repr.plot.height=6)

p = p + facet_wrap(~Species, ncol = 1)

p
```

## facet_grid

```r
mpg %>% head
```

```r
mpg %>% group_by(drv) %>% tally() %>% arrange(desc(n))
```

```r
mpg %>% group_by(cyl) %>% tally() %>% arrange(desc(n))
```

```r
options(repr.plot.width=10, repr.plot.height=8)

p = ggplot(mpg, aes(x = displ, y = hwy))

# geom_point: data will be visualized as points
p = p + geom_point()

# call facet_grid RATHER THAN WRAP --> we specify that we want drv and cyl to be the pairwise combinations 
# for each pairwise combo of drive and cyl, we now have a single chart. 
# ex, the furthermost left box is a 4 cylendar car, and 4 wheel drive --> and we can see displ and hwy for all those combinations 
p = p + facet_grid(drv ~ cyl)

#this will increase the size of the text for the labels
p = p + theme(strip.text = element_text(size = 20))

p
```
