```r
encodings = aes(x = displ, y = hwy)

p = ggplot(mpg, encodings) +  # store the output in the letter p (for plot). mpg is data, where do "hwy" and "displ" come from
    geom_point(size = 2)      # create the scatter plot, where each dot is size 2.

p   # p will now run this chart!
```
