# slice_max

```r
mpg |> write_csv('mpg.csv')
```

```r
top.hwy = mpg |> slice_max(hwy, n = 1)

top.hwy
```

```r
p = ggplot(mpg, aes(x = displ, y = hwy, color = class))

p = p + geom_point()

p = p + geom_text_repel(data = top.hwy, 
                        #use only the top.hw table that we created above - which only includes those two rows, 
                        #so we only show labels for the two top data points 
                        nudge_x = 1,
                        direction = 'y',
                        aes(label = manufacturer), 
                        color = 'black')  # this is important

p.hwy = p

p
```

```r
p = ggplot(mpg, aes(x = displ, y = cty, color = class))

p = p + geom_point()

p = p + geom_text_repel(data = top.hwy, 
                        nudge_x = 1,
                        direction = 'y',
                        aes(label = manufacturer), 
                        color = 'black')

p.cty = p

p
```

```r
c(100, 45, 50, 60) > 50
#we're evaluating whether each value in the c() is greater than 50 -- and we get true or false 
```

```r
sum(c(100, 45, 50, 60) > 50) # tells us how many of these values satisfy this criteria
#false are counted as 0, true are counted as 1
```

```r
#how many times per month, did we see a value above 50? 
weather |>
    group_by(month) |>
    summarize(max_temp = max(temp, na.rm = TRUE),
    # calculate the max_temp for each month 
              #na.rm = TRUE will remove all na values (missing values) when taking the sum
              #if you don't do this, it will override all the values as na - if there are any values of na
              obs = n(), 
              #count the number of observations we observe for each month
              n_above_50 = sum(temp > 50, na.rm = TRUE),
              #count the number of temp values in each month group where the temp is above 50 
              frac_above_50 = sum(temp > 50, na.rm = TRUE) / n())
              #divide by n, the total number of observations, to get a fraction 
```

```r
# let's create a column that shows our median values of temp for each month 
temp_median = weather |> 
    group_by(month) |>
    summarize(temp = median(temp, na.rm = TRUE))
temp_median |> head() 
```

```r
p = ggplot(weather, aes(x = month, y = temp))

p = p + geom_point(alpha = 0.4, 
                   position = position_jitter(h = 0, w = 0.4)) 

#add the median points in red ontop of our graph - using the temp_median data table we created above 
p = p + geom_point(data = temp_median, color = 'firebrick') 

#add cross bar lines around the median points 
p = p + geom_crossbar(data = temp_median, mapping = aes(y = temp, ymin = temp, ymax = temp), 
                      #ymin and ymax encodings for the crossbars 
                      #if ymin and ymax are the same number, these bars will be horizontal. 
                      #if you have different values for ymin and ymax, you will have vertical bars. 
                      color = 'steelblue', fatten = 1, width = 0.8) 
                      #fatten and width apply to the thickness of the bar 

p
```
