4. Add an aesthetic mapping for size to sepal width.

```r
# this will make each species a different color! 
p = ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) + geom_point()

p = p + scale_color_brewer(type = "qual", palette = "Set2")

p
```

5. Create a facets plot where each facet represents one iris species. 
(You may have to resize the output again.)

```r
p = p + facet_wrap(~Species, nrow = 2)
p
```
