This function `get_data` will get the new case counts over time from a given state - it will return data from whatever state you want - as determined below. 

````r
get_data = function(state_code) {
    
    state = state_code  
    
    endpoint = paste0("https://data.cdc.gov/resource/9mfq-cb36.json?state=", state)
    
    # Can you make sense of what's happening here?👇
    df = fromJSON(url(endpoint)) %>% 
        #add a column called date - that's based on submission_date 
        mutate(date = as_date(submission_date),
               new_cases = as.numeric(new_case)) %>%
        #select the columns that we need, arrange by date, and return that information to the user 
        select(date, new_cases, state) %>%
        arrange(date)
    
    return(df)
    
    }
````

````r
# show cases by state - start with NC!

cases_UT = get_data('UT')
cases_OR = get_data('OR')

# now let's stack these two individual tables ontop of each other - rbind will bind together the rows 
cases_combined = rbind(cases_UT,cases_OR)

cases_combined |> head() 

````

We can concatenate these two tables using the `rbind` function. Read the help documentation for `rbind` to see how it works, then concatenate your two tables. 

````r
p = ggplot(cases_combined, aes(x = date, y = new_cases)) + 
    geom_col(width = 1) + 
    facet_wrap(~state, nrow = 2)
p
````
